# Tensorflow2.0 小练习

```python
import tensorflow as tf
import numpy as np
```

## 实现softmax函数

```python
def softmax(x):
    ##########
    '''实现softmax函数，只要求对最后一维归一化，不允许用tf自带的softmax函数'''
    exp_x = tf.exp(x)
    prob_x = exp_x / tf.reduce_sum(exp_x, axis=-1, keepdims=True)
    # 使用numpy
    # x_np = np.array(x)
    # # 减去最大值以保证数值稳定性 ，防止指数爆炸
    # x_max = np.max(x_np, axis=-1, keepdims=True)
    # exp_x = np.exp(x_np - x_max)
    # prob_x = exp_x / np.sum(exp_x, axis=-1, keepdims=True)
    # prob_x = tf.constant(prob_x, dtype=x.dtype)
    ##########
    return prob_x

test_data = np.random.normal(size=[10, 5])
(softmax(test_data).numpy() - tf.nn.softmax(test_data, axis=-1).numpy())**2 <0.0001
```

## 实现sigmoid函数

```python
def sigmoid(x):
    ##########
    '''实现sigmoid函数， 不允许用tf自带的sigmoid函数'''
    prob_x = 1 / (1 + tf.exp(-x))
    # # 使用numpy
    # x_np = np.array(x)
    # prob_x = 1.0 / (1.0 + np.exp(-x_np))
    # prob_x = tf.constant(prob_x, dtype=x.dtype)
    ##########
    return prob_x

test_data = np.random.normal(size=[10, 5])
(sigmoid(test_data).numpy() - tf.nn.sigmoid(test_data).numpy())**2 < 0.0001
```

## 实现 softmax 交叉熵loss函数

```python
def softmax_ce(x, label):
    ##########
    '''实现 softmax 交叉熵loss函数， 不允许用tf自带的softmax_cross_entropy函数'''
    loss = -tf.reduce_mean(tf.reduce_sum(label * tf.math.log(x + 1e-12), axis=-1))
    # # 使用numpy
    # prob_np = np.array(x)
    # label_np = np.array(label)
    # loss = -np.mean(np.sum(label_np * np.log(prob_np + 1e-12), axis=-1))
    # loss = tf.constant(loss, dtype=x.dtype)
    ##########
    return loss

test_data = np.random.normal(size=[10, 5])
prob = tf.nn.softmax(test_data)
label = np.zeros_like(test_data)
label[np.arange(10), np.random.randint(0, 5, size=10)]=1.

((tf.reduce_mean(tf.nn.softmax_cross_entropy_with_logits(label, test_data))
  - softmax_ce(prob, label))**2 < 0.0001).numpy()
```

## 实现 sigmoid 交叉熵loss函数

```python
def sigmoid_ce(x, label):
    ##########
    '''实现 softmax 交叉熵loss函数， 不允许用tf自带的softmax_cross_entropy函数'''
    loss = -tf.reduce_mean(label * tf.math.log(x + 1e-12) + (1 - label) * tf.math.log(1 - x + 1e-12))
    # # 使用numpy
    # prob_np = np.array(x)
    # label_np = np.array(label)
    # loss_array = label_np * np.log(prob_np + 1e-12) + (1.0 - label_np) * np.log(1.0 - prob_np + 1e-12)
    # loss = -np.mean(loss_array)
    # loss = tf.constant(loss, dtype=x.dtype)
    ##########
    return loss

test_data = np.random.normal(size=[10])
prob = tf.nn.sigmoid(test_data)
label = np.random.randint(0, 2, 10).astype(test_data.dtype)
print (label)

((tf.reduce_mean(tf.nn.sigmoid_cross_entropy_with_logits(label, test_data))- sigmoid_ce(prob, label))**2 < 0.0001).numpy()
```
